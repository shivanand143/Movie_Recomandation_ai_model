```python
# Step 1 : Import Libraries
import pandas as pd
```

```python
# Step 2 : Import Data
movies = pd.read_csv('https://raw.githubusercontent.com/YBIFoundation/Dataset/main/Movies%20Recommendation.csv')
```

```python
printprint("First 5 rows of the data:")
print(movies.head())
```

```python
print("\nData Information:")
movies.info()
```

```python
print(movies.describe())
```

```python
# Step 3 : Define Target (y) and Features (X)
y = movies['Movie_Popularity']
X = movies.drop(['Movie_Popularity', 'Movie_ID', 'Movie_Title', 'Movie_Release_Date',
                 'Movie_Homepage', 'Movie_Keywords', 'Movie_Overview',
                 'Movie_Production_House', 'Movie_Production_Country',
                 'Movie_Spoken_Language', 'Movie_Tagline', 'Movie_Cast',
                 'Movie_Crew', 'Movie_Director'], axis=1)
```

```python
#Encoding categorical variables
X = pd.get_dummies(X, columns=['Movie_Genre', 'Movie_Language'], drop_first=True)
```

```python
print("\nFirst 5 rows:")
print(X.head())
```

```python
# Step 4 : Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=2529)
```

```python
# Check the shape of train and test samples
print("\nShapes of Train and Test samples:")
print(f"X_train: {X_train.shape}, X_test: {X_test.shape}, y_train: {y_train.shape}, y_test: {y_test.shape}")
```

```python
#Impute missing values using SimpleImputer
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='most_frequent')
```

```python
# Fit and transform on training data
X_train = imputer.fit_transform(X_train)
```

```python
# Transform test data
X_test = imputer.transform(X_test)
```

```python

# Step 5 : Select Model
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
```

```python
# Step 6 : Train or Fit Model
model.fit(X_train, y_train)
```

```python
# Step 7 : Predict Model
y_pred = model.predict(X_test)
```

```python
print("\nPredicted Values:")
print(y_pred)
```

```python
# Step 8 : Model Accuracy
# Calculate Mean Squared Error and R^2 Score
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
#detailed report
print("\nModel Performance Report")
print("=========================")
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"R² Score: {r2:.4f}")
```

```python
#Displaying actual vs predicted values for the first 10 samples
print("\nActual vs Predicted values for the first 10 samples:")
comparison_df = pd.DataFrame({'Actual': y_test[:10], 'Predicted': y_pred[:10]})
print(comparison_df)
```
